```python
# GRADED FUNCTION: cofi_cost_func
# UNQ_C1

def cofi_cost_func(X, W, b, Y, R, lambda_):
    """
    Returns the cost for the content-based filtering
    Args:
      X (ndarray (num_movies,num_features)): matrix of item features
      W (ndarray (num_users,num_features)) : matrix of user parameters
      b (ndarray (1, num_users)            : vector of user parameters
      Y (ndarray (num_movies,num_users)    : matrix of user ratings of movies
      R (ndarray (num_movies,num_users)    : matrix, where R(i, j) = 1 if the i-th movies was rated by the j-th user
      lambda_ (float): regularization parameter
    Returns:
      J (float) : Cost
    """
    nm, nu = Y.shape
    J = 0
    ### START CODE HERE ###
    for j in range(nu):
        w = W[j,:]
        b_j = b[0,j]
        for i in range(nm):
            x = X[i,:]
            y = Y[i,j]
            r = R[i,j]
            J += r * np.square((np.dot(w,x) + b_j - y ))
    J += (lambda_) * (np.sum(np.square(W)) + np.sum(np.square(X)))
    J = J/2
    ### END CODE HERE ###

    return J
```

```python
import numpy as np

# Función de carga de datos
def load_precalc_params_small():
    file_X = open('small_movies_X.csv', 'r')
    X = np.loadtxt(file_X, delimiter=",")
    file_X.close()

    file_W = open('small_movies_W.csv', 'r')
    W = np.loadtxt(file_W, delimiter=",")
    file_W.close()

    file_b = open('small_movies_b.csv', 'r')
    b = np.loadtxt(file_b, delimiter=",")
    file_b.close()

    return X, W, b

def load_ratings_small():
    file_Y = open('small_movies_Y.csv', 'r')
    Y = np.loadtxt(file_Y, delimiter=",")
    file_Y.close()

    file_R = open('small_movies_R.csv', 'r')
    R = np.loadtxt(file_R, delimiter=",")
    file_R.close()

    return Y, R

# Cargar datos
X, W, b = load_precalc_params_small()
Y, R = load_ratings_small()

# Reducir tamaño del conjunto de datos
num_users_r = 4
num_movies_r = 5
num_features_r = 3

X_r = X[:num_movies_r, :num_features_r]
W_r = W[:num_users_r, :num_features_r]

if b.ndim == 1:
    b_r = b[:num_users_r].reshape(1, -1)
else:
    b_r = b[0, :num_users_r].reshape(1, -1)

if Y.ndim == 1 and R.ndim == 1:
    Y_r = Y[:num_movies_r]
    R_r = R[:num_movies_r]
else:
    Y_r = Y[:num_movies_r, :num_users_r]
    R_r = R[:num_movies_r, :num_users_r]

# Evaluar la función de costo
J = cofi_cost_func(X_r, W_r, b_r, Y_r, R_r, 0)
print(f"Cost: {J:0.2f}")
```

```python
# Evaluate cost function with regularization
J = cofi_cost_func(X_r, W_r, b_r, Y_r, R_r, 1.5);
print(f"Cost (with regularization): {J:0.2f}")
```

```python
def cofi_cost_func_v(X, W, b, Y, R, lambda_):
    """
    Returns the cost for the content-based filtering
    Vectorized for speed. Uses tensorflow operations to be compatible with custom training loop.
    Args:
      X (ndarray (num_movies,num_features)): matrix of item features
      W (ndarray (num_users,num_features)) : matrix of user parameters
      b (ndarray (1, num_users)            : vector of user parameters
      Y (ndarray (num_movies,num_users)    : matrix of user ratings of movies
      R (ndarray (num_movies,num_users)    : matrix, where R(i, j) = 1 if the i-th movies was rated by the j-th user
      lambda_ (float): regularization parameter
    Returns:
      J (float) : Cost
    """
    j = (tf.linalg.matmul(X, tf.transpose(W)) + b - Y)*R
    J = 0.5 * tf.reduce_sum(j**2) + (lambda_/2) * (tf.reduce_sum(X**2) + tf.reduce_sum(W**2))
    return J
```

```python
# Evaluate cost function
J = cofi_cost_func_v(X_r, W_r, b_r, Y_r, R_r, 0);
print(f"Cost: {J:0.2f}")

# Evaluate cost function with regularization
J = cofi_cost_func_v(X_r, W_r, b_r, Y_r, R_r, 1.5);
print(f"Cost (with regularization): {J:0.2f}")
```

```python
import numpy as np

def load_ratings_small():
    # Ejemplo de carga de datos
    Y = np.loadtxt('small_movies_Y.csv', delimiter=",")
    R = np.loadtxt('small_movies_R.csv', delimiter=",")
    return Y, R

def normalizeRatings(Y, R):
    # Función de normalización de ejemplo
    Ymean = np.zeros(Y.shape[0])
    Ynorm = np.zeros(Y.shape)
    for i in range(Y.shape[0]):
        idx = R[i, :] == 1
        if np.any(idx):
            Ymean[i] = np.mean(Y[i, idx])
            Ynorm[i, idx] = Y[i, idx] - Ymean[i]
    return Ynorm, Ymean

# Cargar datos de calificaciones y ajustar tamaño
Y, R = load_ratings_small()

num_movies = Y.shape[0]  # Ajustar según el número de filas en Y
my_ratings = np.zeros(num_movies)

# Asignar calificaciones
my_ratings[2700] = 5
my_ratings[2609] = 2
# ... (otras calificaciones)

# Concatenar nuevas calificaciones con las existentes
Y = np.c_[my_ratings, Y]
R = np.c_[(my_ratings != 0).astype(int), R]

# Normalizar el dataset
Ynorm, Ymean = normalizeRatings(Y, R)

print("Normalized Ratings:", Ynorm)
print("Mean Ratings:", Ymean)
```

```python
#  Useful Values
num_movies, num_users = Y.shape
num_features = 100

# Set Initial Parameters (W, X), use tf.Variable to track these variables
tf.random.set_seed(1234) # for consistent results
W = tf.Variable(tf.random.normal((num_users,  num_features),dtype=tf.float64),  name='W')
X = tf.Variable(tf.random.normal((num_movies, num_features),dtype=tf.float64),  name='X')
b = tf.Variable(tf.random.normal((1,          num_users),   dtype=tf.float64),  name='b')

# Instantiate an optimizer.
optimizer = keras.optimizers.Adam(learning_rate=1e-1)
```

```python
iterations = 200
lambda_ = 1
for iter in range(iterations):
    # Use TensorFlow’s GradientTape
    # to record the operations used to compute the cost
    with tf.GradientTape() as tape:

        # Compute the cost (forward pass included in cost)
        cost_value = cofi_cost_func_v(X, W, b, Ynorm, R, lambda_)

    # Use the gradient tape to automatically retrieve
    # the gradients of the trainable variables with respect to the loss
    grads = tape.gradient( cost_value, [X,W,b] )

    # Run one step of gradient descent by updating
    # the value of the variables to minimize the loss.
    optimizer.apply_gradients( zip(grads, [X,W,b]) )

    # Log periodically.
    if iter % 20 == 0:
        print(f"Training loss at iteration {iter}: {cost_value:0.1f}")
```
